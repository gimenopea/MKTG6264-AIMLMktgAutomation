```python
import pandas as pd

movies_df = pd.read_csv('movies.csv',encoding='ISO-8859-1')
ratings_df = pd.read_csv('ratings.csv',encoding='ISO-8859-1')
```

```python
movies_df.head()
```

```python
ratings_df.head()
```

```python
#filter movies_df and return only MovieID = 2005
goonies_df = movies_df[movies_df['MovieID'] == 2005]

#join goonies_df with ratings_df on MovieID
goonies_df = pd.merge(goonies_df, ratings_df, on='MovieID',how='left')
```

```python
goonies_df = goonies_df.loc[:, ~goonies_df.columns.str.contains('^Unnamed')]
goonies_df
```

### A look at the ratings

```python
import matplotlib.pyplot as plt


goonies_df['Rating'].plot(kind='hist')
plt.axvline(goonies_df['Rating'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.text(goonies_df['Rating'].mean()+0.01, 5, f"Average Rating {round(goonies_df['Rating'].mean(),2)}")
#label the plot
plt.title('Goonies Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
#show the plot
plt.show()
```

### A look at audience age distribution

```python
goonies_df['Age'].plot(kind='hist')

plt.axvline(goonies_df['Age'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.text(goonies_df['Age'].mean()+0.01, 5, f"Average Age {round(goonies_df['Age'].mean(),2)}")
#label the plot
plt.title('Goonies Ratings Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
goonies_df['Age'].value_counts()
```

```python
#average rating for each age group
goonies_df.groupby('Age')['Rating'].mean()
```

### A look at income level

```python
#show value counts for income level
goonies_df['income_level'].value_counts()
```

### A look at gender distribution

```python
#value counts for Gender
goonies_df['Gender'].value_counts()
```

### Summary:

- based on the observed data, ratings for the Goonies is slightly above 3.5 which does make the movie worse nor good. 
- There appears to be only 2 distinct age groups for the ratings and these are the 18 and 35 year olds. 
- The younger crowd rates The Goonies higher than the older crowd.
