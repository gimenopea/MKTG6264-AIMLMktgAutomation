```python
# Get all UserID in ratings_df that have reviewed the goonies
user_ids = ratings[ratings['MovieID'] == 2005]['UserID'].unique()
```

```python
filtered_ratings_df = ratings[ratings['UserID'].isin(user_ids)]
```

```python
joined_df = pd.merge(filtered_ratings_df, movies, on='MovieID')
```

```python
df = joined_df.loc[:, ~joined_df.columns.str.contains('^Unnamed')]
df.head()
```

```python
goonies_reviews = df[df.MovieID == 2005]
```

### Understanding other genres they like

```python
#return the other movies they like and get a sense of the Genre composition
df[df.UserID.isin(goonies_reviews[goonies_reviews.Rating > 4].UserID)]
```

```python
other = df[df.UserID.isin(goonies_reviews[goonies_reviews.Rating > 4].UserID)]
```

```python
other_n = other[['MovieID','Year','Genre']].drop_duplicates()
```

```python
genres_df = other_n['Genre'].str.split('|',expand=True).stack().reset_index(level=1, drop=True).rename('Genre')
genres_df
```

```python
genre_counts = genres_df.value_counts()
print(genre_counts)
```
