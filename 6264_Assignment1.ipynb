```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://stats.idre.ucla.edu/stat/data/binary.csv')
```

### 1. Understanding class distribution for admit (1) and non-admits (0)

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.countplot(x='admit', data=df).set_title('Admit outcomes')
```

Observation: The data seems to be skewed, with a higher number of non-admits compared to admitted outcomes. This imbalance should be taken into account when analyzing results distributions.

### 2. admit outcomes using key features in GRE, GPA and RANK

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes[0,0].set_title('Plot A: GRE scores by admission outcome')
sns.boxplot(ax=axes[0,0], x='admit', y='gre', data=df)

axes[0,1].set_title('Plot B: GPA scores by admission outcome')
sns.boxplot(ax=axes[0,1], x='admit', y='gpa', data=df)

axes[1,0].set_title('Plot C: Rank by admission outcome')
sns.countplot(ax=axes[1,0], x='rank', hue='admit', data=df)

axes[1,1].set_title('Plot D: GRE vs GPA by admission outcome')
sns.scatterplot(ax=axes[1,1], x='gpa', y='gre', hue='admit', data=df);
```

Observations:
Plot A. It can be seen that admitted students have a higher mean GRE scores (above 600) compared to non-admittent students. Both distribution have outliers.

Plot B. Similar to plot A, admitted students have higher mean GPA scores compared to non admit students. 

Plot C: It can be seen that the distribution of admitted students are higher in the higher ranked schools (rank 1 and 2 have higher mean height than rank 3+ schools)

Plot D: It is observed that the distribution of admitted students tends to cluster the higher the GPA alongside the value of their GPA scores. Higher GPA and higher GRE is distributed more to the upper top right region of the plot
